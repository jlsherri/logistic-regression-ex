```python
#filling in null values for age
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else:
        return Age
```

```python
train['Age']=train[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#dropping cabin column, too many missing values
train.drop('Cabin',axis=1,inplace=True)
```

```python
train.head()
```

```python
train.dropna(inplace=True)
```

```python
sns.heatmap(train.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
##missing values all filled in
```

```python
#create dummy variables/dataframes for categorical columns like sex,
#make sure that one column is not a perfect predictor
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=True)
```

```python
sex.head()
```

```python
embark = pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
embark.head()
```

```python
train = pd.concat([train,sex,embark],axis=1)
```

```python
train.head()
```

```python
train.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
train.head()
```

```python
train.drop('PassengerId',axis=1,inplace=True)
```

```python
train.head()
```

```python
X = train.drop('Survived',axis=1)
y = train['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```

```python
#Not terrible accuracy, with 216 correct predictions out of 267, or about 81% overall accuracy.
```
